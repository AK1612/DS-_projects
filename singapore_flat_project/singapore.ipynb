```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df1=pd.read_csv("D:\ds\singapore flat resale  value prediction\ResaleFlatPricesBasedonApprovalDate19901999 (1).csv")
df2=pd.read_csv("D:\ds\singapore flat resale  value prediction\ResaleFlatPricesBasedonApprovalDate2000Feb2012 (1).csv")
df3=pd.read_csv("D:\ds\singapore flat resale  value prediction\ResaleFlatPricesBasedonRegistrationDateFromMar2012toDec2014 (1).csv")
df4=pd.read_csv("D:\ds\singapore flat resale  value prediction\ResaleflatpricesbasedonregistrationdatefromJan2017onwards (1).csv")
df5=pd.read_csv("D:\ds\singapore flat resale  value prediction\ResaleFlatPricesBasedonRegistrationDateFromJan2015toDec2016 (1).csv")
```

```python
da=[df1,df2,df3,df4,df5]
df=pd.concat(da)
df
```

```python
c=['remaining_lease','block', 'street_name']
df=df.drop(c,axis=1)
```

```python
c=df.columns
for i in c:
    print(f"{df[i].unique()}"+'\n\n')
```

```python
df=pd.get_dummies(df,columns=['flat_type', 'town', 'storey_range','flat_model'],dtype='int')
```

```python
df['month']=pd.to_datetime(df['month'],format='%Y-%m').dt.year
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

X=df.drop(['resale_price'],axis=1)
y = df['resale_price']

x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.25)

f=open('regressor_model_comparison.txt','x')

models = [
    LinearRegression(),Lasso(), Ridge(), ElasticNet(),
    KNeighborsRegressor(),
    DecisionTreeRegressor(),
    RandomForestRegressor(),
    GradientBoostingRegressor()
]



for model in models:
    model.fit(x_train,y_train)
    train_prediction = model.predict(x_train)
    test_predicition = model.predict(x_test)

    R2SCORE=r2_score(y_train,train_prediction)
    MAE=mean_absolute_error(y_train,train_prediction)
    MSE=mean_squared_error(y_train,train_prediction)

    mae= mean_absolute_error(y_test,test_predicition)
    mse=mean_squared_error(y_test,test_predicition)
    r2score=r2_score(y_test,test_predicition)
    with open('regressor_model_comparison.txt','a') as f:
        f.write('model =' + repr(model) + '\n')
        f.write('train' + '\n')
        f.write('MAE ='+ repr(MAE) + '\n')
        f.write('MSE ='+ repr(MSE) + '\n')
        f.write('R2 ='+ repr(R2SCORE) + '\n' +'\n')
        f.write('test' + '\n')
        f.write('MAE ='+ repr(mae) + '\n')
        f.write('MSE ='+ repr(mse) + '\n')
        f.write('R2 ='+ repr(r2score) + '\n' +'\n'+'\n')
        


    print(f"******{type(model).__name__}********")
    print(f"********Train*********")
    print(f"MAE: {mean_absolute_error(y_train,train_prediction)}")
    print(f"MSE: {mean_squared_error(y_train,train_prediction)}")
    print(f"R2SCORE: {r2_score(y_train,train_prediction)}")
    print(f"********Test*********")
    print(f"MAE: {mean_absolute_error(y_test,test_predicition)}")
    print(f"MSE: {mean_squared_error(y_test,test_predicition)}")
    print(f"R2SCORE: {r2_score(y_test,test_predicition)}")
    print("\n\n\n")
```
